## Getting Started with OpenCV

```python
import cv2
image=cv2.imread('original.jpg')
```

```python
#Reducing image size to let it fit into the screen properly
image = cv2.resize(image, (600,600))

#Displaying image 
cv2.imshow('image', image)

#To close the image upon pressing 'q'
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Converting RGB image to grayscale:
grey = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
#Writing greyscale image to disk:
cv2.imwrite('grey.jpg', grey)
```

```python
#Converting RGB image to binary 
# - first convert RGB To grey, pass it as first argument to threshold function
# - second argument gives threshold value
# - third argument specifies the value to which pixels having values above threshold should be converted
# - fourth gives the type of thresholding
ret, binary_img = cv2.threshold(grey,127,255,cv2.THRESH_BINARY)

#Displaying the new binary image: 
cv2.imshow('binary', binary_img)
cv2.waitKey(0)
cv2.destroyAllWindows()

#Saving the binary image
cv2.imwrite('binary.jpg', binary_img)
```

## Basics of thresholding

### Simple thresholding 

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Using a shadowed image to study the difference in various thresholding methods:
img = cv2.imread('shadowed.jpeg')
#Displaying image inline using matplotlib
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))   
```

```python
#since thresholding is done only on greyscale images, 
grey = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
#Displaying grey image
plt.imshow(cv2.cvtColor(grey, cv2.COLOR_BGR2RGB)) 
```

```python
#Applying simple binary thresholding
#  --this is the same as the earlier method to convert greyscale image to binary

ret, simple_binary = cv2.threshold(grey, 127, 255, cv2.THRESH_BINARY) 
plt.imshow(cv2.cvtColor(simple_binary, cv2.COLOR_BGR2RGB)) 
```

```python
#Applying simple inverse binary thresholding

ret, simple_binary_inv = cv2.threshold(grey, 127, 255, cv2.THRESH_BINARY_INV) 
plt.imshow(cv2.cvtColor(simple_binary_inv, cv2.COLOR_BGR2RGB)) 
```

```python
#Applying truncated thresholding

ret, trunc_thresh = cv2.threshold(grey, 127, 255, cv2.THRESH_TRUNC) 
plt.imshow(cv2.cvtColor(trunc_thresh, cv2.COLOR_BGR2RGB)) 
```

```python
#Thresholding to Zero and its inverse

ret, thresh2zero = cv2.threshold(grey, 127, 255, cv2.THRESH_TOZERO)
ret, thresh2zero_inv = cv2.threshold(grey, 127, 255, cv2.THRESH_TOZERO_INV) 

plt.imshow(cv2.cvtColor(thresh2zero, cv2.COLOR_BGR2RGB))
plt.show()
plt.imshow(cv2.cvtColor(thresh2zero_inv, cv2.COLOR_BGR2RGB))
plt.show()
```

### Adaptive Thresholding

- the method where threshold value is calculated for smaller regions. 
- leads to different threshold values for different regions (accomodates shadows etc)

The syntax is:
cv2.adaptiveThreshold(source, maxVal, adaptiveMethod, thresholdType, blocksize, constant)

```python
#Applying adaptive thresholding

## Using method cv2.ADAPTIVE_THRESH_GAUSSIAN_C
adaptive_thresh1 = cv2.adaptiveThreshold(grey, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 25, 13)
plt.imshow(cv2.cvtColor(adaptive_thresh1, cv2.COLOR_BGR2RGB))
```

```python
## Using method cv2.ADAPTIVE_THRESH_MEAN_C
adaptive_thresh2 = cv2.adaptiveThreshold(grey, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 25, 13)
plt.imshow(cv2.cvtColor(adaptive_thresh2, cv2.COLOR_BGR2RGB))
```

### Observations:

How to choose optimal value of blocksize and constant?
 - Increasing the blocksize seems to sharpen the image
 - Blocksize can only be odd numbers

## Otsu's Thresholding

References to read up on Otsu's method:
- Paper analysing the method - http://ijarcet.org/wp-content/uploads/IJARCET-VOL-2-ISSUE-2-387-389.pdf
- Detailed mathematical explanation - https://www.cse.unr.edu/~bebis/CS791E/Notes/Thresholding.pdf


```python
img = cv2.imread('shadowed.jpeg')
#since thresholding is done only on greyscale images, 
grey = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
ret,thres = cv2.threshold(grey,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
plt.imshow(cv2.cvtColor(thres, cv2.COLOR_BGR2RGB))
plt.show()
```
